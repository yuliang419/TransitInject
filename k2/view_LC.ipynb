```python
files = glob.glob('mdwarfs/*_ltf.lc')
inds = np.random.randint(0, len(files), 7)

ids = [files[i] for i in inds]
for kic in ids:
    t, f, newf = np.genfromtxt(kic, unpack=True, usecols=(0,1,2))
    fig = plt.figure(figsize=(15,4))
    plt.annotate(kic, xy=(0.8, 0.9), xycoords='axes fraction')
    plt.plot(t,f, 'b.')
    plt.plot(t,newf,'r.')
```

```python
t, mag, newmag = np.genfromtxt('mdwarfs/201665284_ltf.lc', unpack=True, usecols=(0,1,2))
model = LombScargleFast(fit_offset=True).fit(t, newmag)
period, power = model.periodogram_auto(oversampling=100)
```

```python
fig = plt.figure(figsize=(15,6))
plt.plot(t, newmag, 'b.')
# plt.xlim(25,30)
plt.ylim(0.9,1.1)
```

```python
fig = plt.figure(figsize=(15,6))
plt.plot(1./period,power)
plt.xlim(0,60)
plt.ylim(-0.1,1)
plt.ylabel('Power')
plt.xlabel('Period (days)')
```

```python
model.optimizer.period_range = (0.1, 30)
print model.best_period
```

```python
t -= t[0]
ph = t % model.best_period
```

```python
fig = plt.figure(figsize=(15,6))
plt.plot(ph, newmag, 'b.')
plt.ylim(0.9, 1.1)
```
